#### Merge files from Award Data Archive download to create one file for each fiscal year

```python
# Set the directory containing all raw data
rootDir = '/Users/alia/Documents/Github/DoDContractApp/Raw Data'

# Loop through subfolders in directory
for dirName, subdirList, fileList in os.walk(rootDir):
    print('Found directory: %s' % dirName)
    
    # If there are files in the folder
    if len(fileList)>0:
        
        # Create empty dataframe
        df = pd.DataFrame()
        
        for fname in fileList:
            if fname.endswith('.csv'): 
                print('\t%s' % fname)
                filepath = f'{dirName}/{fname}'
                dat = pd.read_csv(filepath)
                df = df.append(dat,ignore_index=True)
        yr = dirName[-6:]
        
        if df.shape[0]>0:
            df.to_csv(f'/Users/alia/Documents/Github/DoDContractApp/Clean Data/Full Year Tables/{yr}_full.csv',index=False)
```
